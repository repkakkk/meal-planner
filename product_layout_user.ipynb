```python
name = 'Хибины, 27 февраля - 7 марта 2021'
sdate = date(2021, 2, 27)   # Дата начала в формате год, месяц, день
edate = date(2021, 3, 7)   # Дата окончания
persons_number = 10        # Число участников
```

```python
c = Cookbook('cookbook_winter.json')
menu = Menu(name, persons_number)
dlist = []
delta = edate - sdate       # as timedelta
locale.setlocale(locale.LC_ALL, 'ru_RU.UTF-8')
for i in range(delta.days + 1):
    day = sdate + timedelta(days=i)
    dlist.append(day.strftime("%-d %B, %A").lower())
d = Day(dlist[0])
d.add_eating(Eating('обед в поезде', contents = [c.find("суп в кружке"), c.get_random("Сладости"), c.find("чай")]))
d.add_eating(Eating('ужин в поезде', contents = [c.find("пюре с карбонадом"), c.get_random("Сладости"), c.find("чай")]))
menu.add_day(d)
for day in dlist[1:-1]:
    d = Day(day)
    get_random_day_winter(d)
    menu.add_day(d)
d = Day(dlist[-1])
d.add_eating(Eating('завтрак в поезде', contents = [c.find("овсяная каша"), c.get_random("Сладости"), 
                                               c.find("кофе"), c.find("чай")]))
d.add_eating(Eating('обед в поезде', contents = [c.find("суп в кружке"), c.get_random("Сладости"), c.find("чай")]))
menu.add_day(d)
menu.to_pdf()
```
